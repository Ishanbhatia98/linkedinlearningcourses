```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
smoking = pd.read_csv('whickham.csv')
```

```python
smoking['ageGroup'] = pd.cut(smoking.age, [0,30,40,53,64], labels=['0-30','30-40','40-53','53-64'])
```

```python
bysmoker = smoking.groupby('smoker').outcome.value_counts(normalize=True)
```

```python
byage = smoking.groupby(['ageGroup','smoker']).outcome.value_counts(normalize=True)
```

```python
plt.subplot(1, 2, 1)
smoking.outcome.value_counts().plot(kind='pie', figsize=(10,5), colors=['red','blue'])
plt.title('outcome')
plt.subplot(1, 2, 2)
smoking.smoker.value_counts().plot(kind='barh', figsize=(10,5), color = ['green', 'orange'])
plt.title('smoker')
```

```python
bysmoker.plot(kind='bar')
```

```python
bysmoker.unstack().plot(kind='bar',stacked=True)
```

```python
byage.unstack().plot(kind='bar', stacked=True)
```

```python
byage.unstack().drop('Dead',axis=1).unstack().plot(kind='bar', stacked=True)
```
