```python
import pandas as pd
import matplotlib.pyplot as plt

from ipywidgets import interact, widgets
```

```python
data = pd.read_csv('gapminder.csv')
```

```python
data.head()
```

```python
data[data.year==1965].plot.scatter('babies_per_woman', 'age5_surviving');
```

```python
def plotyear(year):
    datay = data[data.year==year]
    area = 5e-6*datay.population
    colors = datay.region.map({'Africa':'skyblue', 'Europe':'gold', 'America':'palegreen', 'Asia':'coral'})
    datay.plot.scatter('babies_per_woman', 'age5_surviving', s=area, c=colors, figsize=(12,9), edgecolor='k')
```

```python
plotyear(1965)
```

```python
interact(plotyear, year=widgets.IntSlider(min=1950, max=2015, step=1, value=1965))
```
