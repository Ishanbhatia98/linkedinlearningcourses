```python
data = data.fillna(0)
```

```python
colm = ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu']
colf = ['f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu']
mr, fr = [], []
for c in colm:
    mr.append(int(sum(data[c])))

    
for c in colf:
    fr.append(int(sum(data[c])))
```

```python
sex = ['m'] * len(mr) + ['f'] * len(fr)
r = mr + fr
year  = list(range(1989, 1999)) * 2
country = ['AD'] * len(r)
age = age * 2
```

```python
new = pd.DataFrame({'country':country, 'year':year, 'age':age, 'sex':sex, 'cases':r})
```
